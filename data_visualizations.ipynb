# We plot high dimensional data in R^2 via various techniques to see the differrences:
* PCA
* t-SNE

```python
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=30, centers=3, n_features=100, cluster_std=5, random_state=1)
```

```python
from matplotlib.colors import ListedColormap
cmap = ListedColormap(['#FF0000', '#00FF00', '#0000FF', '#FFFF00', '#FF00FF']).reversed()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=5)
X_pca = pca.fit_transform(X)
```

```python
plt.plot(range(1,6),pca.explained_variance_ratio_,marker='*')
print(pca.explained_variance_ratio_)
```

```python
# We observe that the two componenets are enough to explain most of the data variance
```

```python
X_pca = X_pca[:,0:2]
```

```python
plt.scatter(X_pca[:,0],X_pca[:,1],c=y,cmap=cmap)
plt.title("Visualization of the data using PCA in 2 dims")
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2, random_state=1, perplexity=5)
X_tsne = tsne.fit_transform(X)
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap=cmap)
plt.title("Visualization of the data using t-SNE in 2 dims")
plt.show()
```
